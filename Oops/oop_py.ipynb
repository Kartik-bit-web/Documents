```python
#Inheritance In OOP
#Type of Inherit: Single, Multiple: Hirarchical, Hybrid Inheritance

class Person:
 
    # __init__ is known as the constructor
    def __init__(self, name, id):
        self.name = name
        self.id = id
     
    def display(self):
        print("My name is {}".format(self.name))
        print("IdNumber: {}".format(self.id))
     
# child class
class Employee(Person):
    def __init__(self, name, id, salary):
        self.salary = salary

        #here is we inheritance the class attributes
        Person.__init__(self, name, id)

    def detail(self):
        return f'i am {self.name} My id is {self.id} and salary is {self.salary}'
    
x = Employee('Kartik', 1, 20000)
print(x.display())
print(x.detail())
 
```

```python
#Inherit Multi-Inheritance-->
#Here we can call it Multi Parent inheritance 

#this is first Parent 
class Base1:
    def __init__(self):
        self.a = 'kartik'

#this is Second Parent 
class Base2:
    def __init__(self):
        self.b = 'Mehra'

#this is Chile Who inherit from Both Parents
class Both(Base1, Base2):
    def __init__(self):
        Base1.__init__(self)
        Base2.__init__(self)

    def printBoth(self):
        return f'firstName is {self.a} lastName is {self.b}'

obj = Both()
print(obj.printBoth())
```

```python
#Inheritance MultiLevel-->
class G_prent:
    def __init__(self):
        self.name = 'Tholo Ram'
    
    def getName(self):
        return self.name

class Perent(G_prent):
    def __init__(self):
        self.p_name = 'Jugal Kishore'
        G_prent.__init__(self)

    def gp_name(self):
        return f'i am {self.p_name} father name is {self.name}'
    
class child(Perent):
    def __init__(self, c_name):
        Perent.__init__(self)
        self.c_name = c_name
    
    def c_call(self):
        return f'i am {self.c_name} and  father name is {self.p_name} and grandFather name is {self.name}'

#Here is the Idea
acc = child('Kartik')

#this method inner child class
print(acc.c_call())

#this method inner Perent class
print(acc.gp_name())

#this method inner GrandFaher class
print(acc.getName())

#Everything we can access through acc Object of Child class
 
```

```python
#Polymorphism-->
#It's mean having many forms. 
#eg:We need to determine if the gives species of birds fly or no

class Bird:
    def intro(self):
        print('There are many type of birds')

    def fly(self):
        print("Most is birds can fly but some cannot")

class Sparrow(Bird):
    def flight(self):
        print("Sparrow can fly")

class Ostrich(Bird):
    def flight(self):
        print("Ostrich cannot fly")

obj_spa = Sparrow()
obj_ost = Ostrich()

#this inherit intro and fly from Bird
obj_spa.intro()
#this inherit from sparrow class
obj_spa.flight()
# this again Bird class
obj_spa.fly()
#this inherit from Orsrich
obj_ost.flight()
```

```python
#Encapsulation-->
#It is one of fundamental concept in OOP.
#It describe the idea and method that work on dara within one 
#unit
#to pervent accidental change , an object variablecan be changed
#an object's method. those are variable are Private variables
class Base:
    def __init__(self):
        self.a = "lol"
        #here is private variable cannot access from outter
        self.__c = "kartik"

    #But by doing this we can share it info
    def grav(self):
        print(self.__c)

class Derived(Base):
    def __init__(self):
        Base.__init__(self)
        print("calling private variable")
        print(self.__c)


y = Base()
#If we Directly want to see variable then we cannot access private variable
print(y.a)

#doing method this then we can acess
print(y.grav())

#print(y.__c)
```
